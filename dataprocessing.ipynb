```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("archive/cbb.csv")

df.head()
df.info()
df.describe()

plt.figure(figsize=(12, 10))  # Bigger plot
sns.heatmap(
    df.select_dtypes(include='number').corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True,
    cbar=True
)

# Plot histogram first
plt.figure(figsize=(8, 4))
sns.histplot(df['W'], bins=20, kde=True)
plt.title("Wins per Team")
plt.xlabel("Wins")
plt.ylabel("Frequency")
plt.show()

# Then the cleaned-up heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
